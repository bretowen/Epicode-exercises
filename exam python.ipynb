```python
#filtro il df in modo da non avere valori nulli nella colonna "continent"
continent_cases= df[df["continent"].notna()] 

#seleziono ultima data di ogni paese (location). Come? prima ordino per valori appena trovati (continenti senza nulli), poi raggruppo per location, e poi -essendo tutto in ordine crescente- chiedo l'ultimo valore, che corrisponde al numero finale cumulativo. 
latest_data= continent_cases.sort_values("date").groupby("location").tail(1) 

# Raggruppiamo i casi per continente e sommiamo i casi totali
cases_by_continent= latest_data.groupby("continent")["total_cases"].sum()
cases_by_continent
```

```python
#esercizio n.2.b (percentuale per continente rispetto al numero totale di casi)
#per prima cosa devo calcolare il numero totale di casi
#poi faccio l'operazione matematica per capire il rapporto (=percentuale)
```

```python
world_total_cases= cases_by_continent.sum()
percentage_by_continent= (cases_by_continent / world_total_cases)*100
percentage_by_continent.round(2)
```

```python
#esercizio n.3
#Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente, filtrare via i giorni che non hanno misurazioni; 

#quindi mostrare con dei grafici adeguati:
    #l'evoluzione del casi totali dall'inizio alla fine dell'anno
    #il numero di nuovi casi rispetto alla data
```

```python
#per prima cosa converto il "date" in "datetime"
#poi opero due ulteriori filtri che mi restituiscono i valori per la sola location "Italy" e per il solo anno 2022.  
#infine, al risultato precedentemente ottenuto coi due filtri, ho operato un ulteriore filtro che mi seleziona SOLO i new cases che non sono nulli & i new cases che sono diversi da 0

df["date"] = pd.to_datetime(df["date"])  #la data era letta automaticamente come stringa e non mi permette certi calcoli (es. filtrarle o ordinarle o estrarre il mese. etc.). Devo allora convertirla.
Italy_2022 = df[(df["location"] == "Italy") & (df["date"].dt.year == 2022)] #ho operato 2 filtri che mi restituiscono i valori per la sola location "Italy" e per il solo anno 2022 
Italy_2022 = Italy_2022[Italy_2022["new_cases"].notna() & (Italy_2022["new_cases"]!=0)]  #del risultato dei filtri ho operato un ulteriore filtro che mi seleziona tutti i dati dell'Italia con i new_cases non-nulli & i new cases diversi da 0
Italy_2022
```

```python
#esercizio 3
#Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente, filtrare via i giorni che non hanno misurazioni; quindi mostrare con dei grafici adeguati:
#a: l'evoluzione del casi totali dall'inizio alla fine dell'anno
#b: il numero di nuovi casi rispetto alla data
```

```python
#a

import matplotlib.pyplot as plt
Italy_2022.set_index('date', inplace=True)
Italy_2022['total_cases'].plot(
    figsize=(14, 6),
    title='Casi totali COVID-19 in Italia nel 2022)',
    xlabel='Data',
    ylabel='Casi totali (Milioni)',
    color='royalblue',
    linewidth=2,
    grid=True
)
```

```python
#b

Italy_2022.index = Italy_2022.index.date  #ho convertito da 'datetime' a 'date'
Italy_2022['new_cases'].plot(
    kind='bar',
    figsize=(14, 6),
    title='Nuovi casi COVID-19 in Italia nel 2022)',
    xlabel='Data',
    ylabel='Nuovi casi',
    color='darkorange'
)
```

```python
#4a.

import seaborn as sb
#definisco il periodo temporale di mio interesse (maggio22-apr23)
start_date = "2022-05-01"
end_date = "2023-04-30"

#applico filtro su location + date + icu (solo i valori non nulli)
icu_df = df[
    (df['location'].isin(['Italy', 'Germany', 'France'])) &
    (df['date'] >= start_date) &
    (df['date'] <= end_date) &
    (df['icu_patients'].notna())
]
```

```python
plt.figure(figsize=(10, 6))
sb.boxplot(data=icu_df, x='location', y='icu_patients', palette='pastel')
plt.title('Distribuzione dei pazienti in terapia intensiva (ICU)\nMaggio 2022 - Aprile 2023')
plt.xlabel('Paese')
plt.ylabel('Pazienti in terapia intensiva')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  4b. la Germania ha avuto un numero mediano di pazienti in terapia intensiva leggermente più elevato rispetto alla Francia e notevolmente più elevato rispetto all'Italia. La Germania presenta anche una variabilità maggiore. L’Italia mostra valori più bassi e meno dispersi.
```

```python
# 5a. mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna (colonna hosp_patients) 
countries = ['Italy', 'Germany', 'France', 'Spain'] #ho selezionato i paesi richiesti
df_2021 = df[
    (df['location'].isin(countries)) &
    (df['date'].dt.year == 2021) #ho selezionato anche l'anno richiesto
]
df_hospitalized = df_2021[['date', 'location', 'hosp_patients']]  #filtro il df in modo da ottenere solo le colonne che mi interessano: data, luogo, numero ospedalizzati.

hospitalized_sum= df_hospitalized.groupby("location")["hosp_patients"].sum()  #del df filtrato raggruppo per paese e chiedo la somma degli ospedalizzati
hospitalized_sum
```

```python
plt.figure(figsize=(8, 5))
sb.barplot(x=hospitalized_sum.index, y=hospitalized_sum.values, palette="Set2")
plt.title("Totale pazienti ospedalizzati nel 2021")
plt.ylabel("Somma pazienti ospedalizzati (Milioni)")
plt.xlabel("Paese")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
null_counts = df_hospitalized[df_hospitalized['hosp_patients'].isna()].groupby('location').size()
print("Valori nulli per Paese:")
print (null_counts)
```
